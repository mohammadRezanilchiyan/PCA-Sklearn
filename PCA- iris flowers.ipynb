```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the features (important for PCA)
X_standardized = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Apply PCA to reduce to 2 components (for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standardized)

# Plot the original and reduced data
plt.figure(figsize=(10, 5))

# Original Data
plt.subplot(1, 2, 1)
plt.scatter(X_standardized[:, 0], X_standardized[:, 1], c=y, cmap='viridis')
plt.title('Original Data')
plt.xlabel('Sepal Length (Standardized)')
plt.ylabel('Sepal Width (Standardized)')

# PCA-transformed Data
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

The resulting plot shows how PCA transforms the original high-dimensional data into a lower-dimensional space, capturing most of the variability. The principal components represent new axes along which the data is projected, facilitating visualization and analysis.
